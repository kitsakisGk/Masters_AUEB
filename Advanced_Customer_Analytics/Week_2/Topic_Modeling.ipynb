```python
def tokenizer(doc,sw):
    return [word for word in [re.sub('[^a-z]','',x.lower()) for x in doc.strip().split()] if word not in sw and len(word)>2]
```

```python
import nltk
nltk.download('stopwords')
```

```python
#stopwords
sw=stopwords.words('english')
```

```python
#new LDA model
lda = tp.LDAModel(k=20)

#add the docs to the model
with open('/content/sample_data/news.txt') as f:
    for doc in f:
        lda.add_doc(tokenizer(doc,sw))


#train LDA model
for i in range(0, 500,10):
    lda.train(10)
    print('Iteration: {}\tLog-likelihood: {}'.format(i, lda.ll_per_word))
```

```python
#print doc info
for doc in lda.docs:
    print ([round(p,2) for p in doc.get_topic_dist()])
    print()
```

```python
#print topic info
for k in range(lda.k):

    topk_words=[pair[0] for pair in lda.get_topic_words(k, top_n=15)]
    print(k,topk_words)
    print()

```

```python
#new LDA model
ct = tp.CTModel(k=20)

#add the docs to the model
with open('/content/sample_data/news.txt') as f:
    for doc in f:
        ct.add_doc(tokenizer(doc,sw))


#train LDA model
for i in range(0, 500,10):
    ct.train(10)
    print('Iteration: {}\tLog-likelihood: {}'.format(i, ct.ll_per_word))
```

```python
#print topic info
for k in range(ct.k):

    topk_words=[pair[0] for pair in ct.get_topic_words(k, top_n=15)]
    print(k,topk_words)

    #get correlations
    corrs=ct.get_correlations(k)
    for i in range(20):
        if i!=k and corrs[i]>0.3:
            print('Related Topic:{} ({:.2f})'.format(i,corrs[i]))

    print()
```
