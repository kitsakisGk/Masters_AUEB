```python
# HTML tutorial: https://www.w3schools.com/html/default.asp
from bs4 import BeautifulSoup
import re
import time
import requests
import csv
```

```python

def get_reviews(
    url:str,
    outpath:str
    ):

    fw=open(outpath,'w',encoding='utf8') # output file

    writer=csv.writer(fw,lineterminator='\n')#create a csv writer for this file

    writer.writerow(['critic','text','outlet','date','rating'])


    html=None


    for i in range(5): # try 5 times

        #send a request to access the url
        response=requests.get(url,headers = { 'User-Agent': 'Mozilla/5.0 (Windows NT 6.1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/41.0.2228.0 Safari/537.36', })
        if response: # explanation on response codes: https://realpython.com/python-requests/#status-codes
            break # we got the file, break the loop
        else:
          time.sleep(2) # wait 2 secs


    # all five attempts failed, return  None
    if not response: return None

    html=response.text# read in the text from the file

    soup = BeautifulSoup(html,'html') # parse the html

    reviews=soup.findAll('div', {'class':re.compile('review-row')}) # get all the review divs

    print(len(reviews))
    for review in reviews:

        critic,text,outlet,date,rating='NA','NA','NA','NA','NA' # initialize key attributes

        #critic
        criticBox=review.find('a',{'href':re.compile('/critics/')})
        if criticBox: critic=criticBox.text.strip()

        #review text
        textBox=review.find('p',{'data-qa':'review-quote'})
        if textBox: text=textBox.text.strip()

        #outlet
        outletBox=review.find('a',{'data-qa':'review-publication'})
        if outletBox: outlet=outletBox.text.strip()

        #date
        dateBox=review.find('span',{'data-qa':'review-date'})
        if dateBox: date=dateBox.text.strip()


        #rating
        ratingBox=review.find('score-icon-critics')
        if ratingBox:
            rating=ratingBox['sentiment']

        writer.writerow([critic,text,outlet,date,rating]) # write to file

    fw.close()


```

```python
url='https://www.rottentomatoes.com/m/space_jam/reviews/'
outpath='space_jam_reviews.csv'
get_reviews(url,outpath)
```
