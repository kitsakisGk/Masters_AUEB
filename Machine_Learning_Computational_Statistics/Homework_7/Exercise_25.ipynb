```python
import scipy.io
import numpy as np
from sklearn.mixture import GaussianMixture
```

```python
# Load the dataset
mat_data = scipy.io.loadmat('Dataset.mat')
X1 = mat_data['X1']  
X2 = mat_data['X2']  
```

```python
# Define test points
points_X1 = np.array([[2.01, 2.99, 3.98, 5.02],
                       [20.78, -15.26, 19.38, -25.02],
                       [3.08, 3.88, 4.15, 6.02]])

points_X2 = np.array([[0.05, 0.15, -0.12, -0.08],
                       [7.18, 7.98, 9.12, 9.94],
                       [3.48, 4.01, 4.55, 4.96],
                       [20.78, -15.26, 19.38, -25.02]])
```

```python
# Fit Gaussian Mixture Model for each dataset
gmm_X1 = GaussianMixture(n_components=2, covariance_type='full', random_state=0)
gmm_X2 = GaussianMixture(n_components=2, covariance_type='full', random_state=0)

gmm_X1.fit(X1)
gmm_X2.fit(X2)
```

```python
# Estimate densities at given points
pdf_X1 = np.exp(gmm_X1.score_samples(points_X1))
pdf_X2 = np.exp(gmm_X2.score_samples(points_X2))
```

```python
# Print results
print("Estimated densities for X1 at given points:")
print(pdf_X1)
print("\nEstimated densities for X2 at given points:")
print(pdf_X2)
```
